```python
def draw_box(df, x_col):
  for col in target_cols:
    plt.figure(figsize=(12, 4))
    sns.boxplot(data=df, x=x_col, y=col)
    plt.title(f'{x_col} 별 {col} 결과')
    plt.xticks(rotation=90)
    # plt.ylim([0, 100])
    plt.show()
```


# 인덱싱

```python
loc_cols = [col for col in df_체급비.columns if '배식장소' in col]
```

```python
df_체급비[loc_cols]
```

```python
for col in loc_cols:
  df_체급비[col].fillna('', inplace=True)

df_체급비[loc_cols[0]].replace('○', '식당', inplace=True)
df_체급비[loc_cols[1]].replace('○', '교실', inplace=True)
df_체급비[loc_cols[2]].replace('○', '식당+교실', inplace=True)
df_체급비[loc_cols[3]].replace('○', '기타', inplace=True)
```

```python
df_체급비[loc_cols]
```

```python
from tqdm import tqdm

df_체급비['배식장소'] = ''

for idx in tqdm(df_체급비.index):
  df_체급비.loc[idx, '배식장소'] = df_체급비.loc[idx, loc_cols].sum()
```

```python
df_체급비['배식장소'].value_counts()
```

```python
조건1 = df_체급비['배식장소'] != ''
조건2 = df_체급비['배식장소'] != '기타'

draw_box(df_체급비[(조건1) & (조건2)], '배식장소')
```

```python
import requests

# API 키와 URL 설정
api_key = '6bfdc02471c4783636d96371c9d25566'
svc_type = 'api'  # 예시로 'api'를 사용했으나 실제로는 문서를 확인하여 올바른 값을 사용해야 합니다.
svcCode = 'COSE_VIEW'
seq=1
url = f'https://www.career.go.kr/cnet/openapi/getOpenApi.json?apiKey={api_key}&svcType={svc_type}&svcCode={svcCode}&seq={seq}'

# API 호출
response = requests.get(url)

# 응답 상태 코드 확인
if response.status_code == 200:
    try:
        # JSON 데이터 파싱
        data = response.json()
        # 데이터를 확인하거나 처리
        print(data)
    except requests.JSONDecodeError:
        # JSON 디코딩 에러 발생 시 응답 텍스트 출력
        print("Failed to decode JSON response")
        print(response.text)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```
